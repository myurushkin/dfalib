```python
ctx = dafna.Context()
pattern = ctx.create_pattern("aa+")
```

```python
list(pattern.min_strings())
```

```python
patterns = []
for i in range(1, 4):
    ggg = 'g' * i
    patterns.append(ctx.create_pattern(ggg + "a" + ggg))
result = dafna.psum(patterns)
```

```python
list(result.min_strings())
```
